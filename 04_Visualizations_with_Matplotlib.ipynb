# OOP Interface

```python
fig, axes = plt.subplots(figsize=[12,6]) 
# Retorna uma tupla que atribui em ambas variáveis fig, axes:
# fig = plt.figure()
# ax = fig.add_subplot(111)
# https://stackoverflow.com/questions/34162443/why-do-many-examples-use-fig-ax-plt-subplots-in-matplotlib-pyplot-python
```

```python
#axes1
axes.plot(
    x,(x ** 2), color = 'red',linewidth=3,
    marker='o', markersize=8, label='X^2')

#axes2
axes.plot(x,-1*(x ** 2),'b--',label='-x^2',color = 'orange')

#Elementos configuráveis de um gráfico
axes.set_xlabel('X', color='pink') # rótulo x
axes.set_ylabel('X Squared',color='blue') # rótulo y
axes.set_title('My Nice Plot') # título
axes.legend() # legenda

fig # retorna o arquvio png de figura completa
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted');

axes.set_title("My Nice Plot")
```

## Other Types of plots
Quando chamamos a função subplot() obtemos uma tupla contendo o elemento figure e o elemento axes

```python
plot_objects = plt.subplots()

fig, ax = plot_objects

ax.plot([1,2,3], [1,2,3])

plot_objects
```

Podemos definir quantos axes podemos ter dentro da figura. Para isso usamos os parametros de nwows e ncols

```python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

#fig, ((ax1, ax2), (ax3, ax4)) = plot_objects

plot_objects
```

```python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6)) # criando a estrutura da figura com duas linhas e duas colunas

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects # estruturando a ordem dos axes sendo (linha 1 e coluna1, linha1 e coluna 2)

plot_objects
```

```python
#cada axe (gráfico) recebe os dados e podemos utilizar atributos da forma do gráfico
ax4.plot(np.random.randn(50), c='yellow')
ax1.plot(np.random.randn(50), c='red', linestyle='--',label='ax1')
ax2.plot(np.random.randn(50), c='green', linestyle=':',label='ax2')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0,label='ax3')


fig
```
